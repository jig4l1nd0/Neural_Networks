### Mtricas de evaluacin para un modelo de clasificacin

Las mtricas de evaluacin se utilizan para evaluar el desempeo de un modelo clasificatorio sobre un conjunto de datos de prueba. Al tratarse de modelos de clasificacin es indispensable contar con al menos una variable independiente de tipo categrico dentro del conjunto de datos (algoritmos supervisados), dicha variable fungir como una etiqueta que nos permitir construir el modelo a partir del conjunto de entrenamiento (calibrar pesos) y adems conocer la clasificacin verdadera de los datos del conjunto de prueba para as poder calcular dichas mtricas de evaluacin.

#### Matriz de confusin

Estas mtricas se derivan de un arreglo tabular conocido como matriz de confusin. Dado un conjunto de datos y considerando una variable categrica de referencia con K posibles clases, niveles o categoras, definimos a la matriz de confusin como el arreglo tabular de K x K donde se contabiliza a las observaciones del conjunto de prueba haciendo referencia  en las columnas a la clasificacin real de las mismas y en las filas a la clasificacin predicha para stas. 

Ej. Supongamos que deseamos conocer las mtricas de evaluacin para la prueba rpida de deteccin del Coronavirus-2019, dicha prueba de deteccin puede pensarse como un modelo clasificatorio que determina de manera rpida si la persona est enferma o no. Sin embargo, la nica manera confiable de detectar al virus es mediante otro tipo de tcnicas que toman ms tiempo, a las cuales denominaremos como prueba lenta. Por lo tanto dada una poblacin de N individuos que pueden caer en una de dos categoras (Infectado, Sano), tenemos la siguiente matriz de confusin hipottica:

Prueba rpida\Prueba lenta | Infectado | Sano
--- | --- | ---
**Infectado** | n_11 | n_12
**Sano** | n_21 | n_22

En general tenemos entonces que el diseo de cualquier matriz de confusin debera de ser el siguiente:

Prediccin\Realidad | Categora_1 | Categora_2 | ... | Categora K
--- | --- | --- | --- | ---
Categoria_1 | n_11 | n_12 | ... | n_1K
Categoria_2 | n_21 | n_22 | ... | n_2K
... | ... | ... | ... | ...
Categoria_K | n_K1 | n_K2 | ... | n_KK

<br></br>
Donde se cumple que:
<br></br>
<div align="center">$N = \sum_{i=1}^{k}\sum_{j=1}^{k} n_{ij}$</div>
<br></br>
y cada $n_{ij}$ representa el conteo de las observaciones que cayeron en la categora $i$ y en la categora $j$.

#### Respuesta

1. En 2016 se estim la prevalencia de VIH en Mxico en la poblacin enre 15 y 49 aos en 0.3%. Sin embargo, dicha prevalencia se encuentra concentrada en los grupos de riesgo: hombres que tienen sexo con hombres, usuarios de drogas intravenosas y trabajadores sexuales. Fuente: https://en.wikipedia.org/wiki/HIV/AIDS_in_Mexico

2. De acuerdo con: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC88179/ , la Sensibilidad de una ELISA de 4a generacin es del 99.5%

3. De acuerdo con: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC88179/ , la Especificidad de una ELISA de 4a generacin es del 99.7%

4. Utilizando el teorema de Bayes, esta pregunta puede plantearse de la siguiente manera:

Consideremos la siguiente notacin:
<br></br>
* $E$: El individuo realmente tiene VIH.
* $E^c$: El individuo realmente NO tiene VIH.
* $P$: La prueba ELISA da Positiva al realizarse, sugiriendo que el individuo tiene VIH.
* $P^c$: La prueba ELISA da Negativa al realizarse, sugiriendo que el individuo NO tiene VIH.
<br></br>
<div align="center">$P(E|P) = \frac {P(E\cap P)}{P(P)} = {{P(P | E) P(E)} \over { P(P\cap E^c) + P(P\cap E)}} = {{P(P | E) P(E)} \over { P(P | E^c) P(E^c) + P(P | E) P(E)}}~~~~~(1)$</div>
<br></br>
Ahora bien, consideremos que la sensibilidad se define como el nmero de individuos diagnosticados Positivo que en realidad presentaban la condicin
<br></br>
<div align="center">$ Sensibilidad = P(P | E) = 0.995$</div>
<br></br>
mientras que la especificidad se definira como el nmero de individuos diagnosticados Negativo
<br></br>
<div align="center">$ Especificidad = P(P^c | E^c) = 0.997$</div>
<br></br>
si y slo si
<br></br>
<div align="center">$ P(P | E^c) = 1 - Especificidad = 1 - P(P^c | E^c) = 0.003$</div>
<br></br>
por otra parte, la prevalencia de la enfermedad en el pas es:
<br></br>
<div align="center">$ Prevalencia = P(E) = 0.003$</div>
<br></br>
por lo que
<br></br>
<div align="center">$ P(E^c) = 1 - P(E) = 0.997$</div>
<br></br>
con estas tres cantidades podemos sustituir en $(1)$:
<br></br>
<div align="center">$P(E|P) = {{(0.995)(0.003)} \over { (0.003)(0.997) + (0.995)(0.003)}} = 0.499498~~~~~(2)$</div>
<br></br>
por lo tanto podemos concluir que la probabilidad de que un mexicano cualquiera est enfermo dado que vivi una situacin de riesgo y su prueba sali positiva es casi de 0.5, lo cual sugiere que se requiere de ms evidencia para concluir que verdaderamente tiene VIH. Este es un fenmeno que variar de pas a pas en funcin de la prevalencia de la enfermedad en cada territorio, por ejemplo, si en lugar de considerar a una persona en Mxico consideramos a una persona en algn pas de frica Oriental o frica Central como Swazilandia, cuya prevalencia de VIH para 2014 ronda el 27.73% ( https://en.wikipedia.org/wiki/Epidemiology_of_HIV/AIDS ) tendramos un clculo bastante diferente:
<br></br>
<div align="center">$P(E|P) = {{(0.995)(0.2773)} \over { (0.003)(0.7227) + (0.995)(0.2773)}} = 0.9922034~~~~~(2)$</div>
<br></br>
es importante notar que un detalle importante para poder calcular lo anterior radica en condicionar tanto especificidad como sensibilidad respecto a la prevalencia poblacional, la cual es un dato que casi siempre podremos conocer.
